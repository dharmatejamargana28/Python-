```python
# Function to calculate total revenue
def calculate_total_revenue(transactions):
    # Sum all the values in the transactions list to get the total revenue
    total_revenue = sum(transactions)
    return total_revenue

# Example list of transactions (amounts in dollars, for example)
transactions = [100, 250, 300, 150, 400, 50, 200]

# Calculate and print the total revenue
total_revenue = calculate_total_revenue(transactions)
print(f"Total Revenue Generated: ${total_revenue}")
```

```python
from collections import Counter

# Function to identify the most frequently purchased product
def most_frequently_purchased(products):
    # Count the frequency of each product in the list
    product_counts = Counter(products)

    # Find the product with the maximum frequency
    most_frequent_product = product_counts.most_common(1)

    if most_frequent_product:
        return most_frequent_product[0]  # (Product, Frequency)
    else:
        return None

# Example list of products purchased (this can be changed to your actual list)
purchased_products = ["Laptop", "Smartphone", "Tablet", "Smartphone", "Laptop", "Smartwatch", "Smartphone"]

# Find and print the most frequently purchased product
result = most_frequently_purchased(purchased_products)
if result:
    print(f"The most frequently purchased product is: {result[0]} with {result[1]} purchases.")
else:
    print("No products purchased.")
```

```python
# Function to calculate average transaction amount
def calculate_average_transaction(transactions):
    # Check if the list is not empty to avoid division by zero
    if len(transactions) == 0:
        return 0
    # Calculate the sum of all transactions and divide by the number of transactions
    average_transaction = sum(transactions) / len(transactions)
    return average_transaction

# Example list of transaction amounts (in dollars)
transactions = [100, 250, 300, 150, 400, 50, 200]

# Calculate and print the average transaction amount
average_transaction = calculate_average_transaction(transactions)
print(f"Average Transaction Amount: ${average_transaction:.2f}")
```

```python
# Function to filter transactions greater than a specified amount
def filter_transactions(transactions, threshold):
    # Use list comprehension to filter transactions greater than the threshold
    filtered_transactions = [amount for amount in transactions if amount > threshold]
    return filtered_transactions

# Example list of transaction amounts (in dollars)
transactions = [100, 250, 300, 150, 400, 50, 200]

# Set the threshold value
threshold = 200

# Filter transactions and print the results
filtered_transactions = filter_transactions(transactions, threshold)
print(f"Transactions greater than ${threshold}: {filtered_transactions}")
```

```python
# Function to find the top 3 customers who have spent the most
def top_3_customers(customer_spending):
    # Sort the customers based on spending in descending order and get the top 3
    sorted_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)
    top_3 = sorted_customers[:3]
    return top_3

# Example data: customer names and their total spending (in dollars)
customer_spending = {
    "Alice": 1200,
    "Bob": 1500,
    "Charlie": 800,
    "David": 2000,
    "Eva": 950
}

# Find and print the top 3 customers who have spent the most
top_3 = top_3_customers(customer_spending)
print("Top 3 customers who have spent the most:")
for customer, spending in top_3:
    print(f"{customer}: ${spending}")
```
